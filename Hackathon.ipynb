## Dropping all the unnecessary Columns

```python
food_df.drop((food_df.columns[16:4159:1]),axis=1,inplace=True)
```

```python
food_df.drop(['address','Mobile','dish_liked','reviews_list','menu_item',],axis=1,inplace=True)
food_df.head()
```

```python
food_df['approx_cost(for two people)']=food_df['approx_cost(for two people)'].replace('\?','',regex=True)
```

```python
food_df.info()
```

### Changing the type of columns which are numerical but showing as object.

```python
food_df["no_of_votes"]=pd.to_numeric(food_df["no_of_votes"],errors='coerce')
```

```python
food_df["approx_cost(for two people)"]=pd.to_numeric(food_df['approx_cost(for two people)'],errors='coerce')
```

```python
food_df.info()
```

## Dropping Duplicates

```python
food_df.drop_duplicates(inplace=True)
food_df.shape
```

```python
food_df.info()
```

## Cleaning Rate Column

```python
food_df['rate'].unique()
```

```python
def handlerate(value):
    if(value== " the food is delish.')") or (value=='do try the Tempura Calamari') or (value=="but for the delivery boy. \\n\\nHave my money refunded or there will be consequences. Zomato and Crunch Pizza"):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return value
    
food_df['rate'] = food_df['rate'].apply(handlerate)
food_df['rate'].head()
```

```python
food_df['rate']=pd.to_numeric(food_df['rate'],errors='coerce')
```

## Filling null Values in rate column

```python
food_df['rate'].fillna(food_df['rate'].mean(), inplace = True)
food_df['rate'].isnull().sum()
```

```python
food_df.info()
```

## Dropping Null Values

```python
food_df.dropna(inplace = True)
food_df.head()
```

## Renaming long Column names

```python
food_df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type','listed_in(city)':'City'}, inplace = True)
food_df.head()
```

```python
food_df.location.unique()
```

### Hence,we have both city and location column so its better to keep only one.

```python
food_df['Cost2plates'].unique()
```

## cleaning rest_type column

```python
rest_types = food_df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

## Marking rest_type less than 1000 as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
food_df['rest_type'] = food_df['rest_type'].apply(handle_rest_type)
food_df['rest_type'].value_counts()
```

## Handling location column

```python
food_df['location'].value_counts(ascending  = False)
```

```python
location = food_df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
food_df['location'] = food_df['location'].apply(handle_location)
food_df['location'].value_counts()
```

## Handling cuisine column

```python
cuisines = food_df['cuisines'].value_counts(ascending  = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
food_df['cuisines'] = food_df['cuisines'].apply(handle_cuisines)
food_df['cuisines'].value_counts()
```

# Visualization

## 1.At first, let's see how many restaurants are there in particular location.

```python
sns.set_style("darkgrid")
```

```python
plt.figure(figsize = (16,10))
ax = sns.countplot(food_df['location'])
plt.xticks(rotation=90)
plt.show()
```

### From the above countplot we can say that BTM has already so many restaurant so if we want to open a restaurant we should not open it in the areas like BTM which already has so many restaurants, If we want to open a restaurant then we should open it in areas like sarjapur road,Ejiipura which has less number of restaurants.

## 2.Let's take a look at how many restaurants have online order facility.

```python
plt.figure(figsize = (6,6))
sns.countplot(food_df['online_order'])
plt.show()
```

### It shows that most of the restaurants have online order facility while few of them doesn't have this facility

## 3. Let's see how many restaurants have booking table facility.

```python
plt.figure(figsize = (6,6))
sns.countplot(food_df['book_table'])
plt.show()
```

### Above graph shows that most restaurants doesn't have Table booking facilty.

## 4. Now lets compare the rating of the restaurants with their online order and booking table facility.

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = food_df)
plt.show()
```

### From the above graph we can say that the restaurants who have online order facility their max rating is bit higher than the restaurants who doesn't have online order facility.

## Book Table vs Rate

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'book_table', y = 'rate', data = food_df)
plt.show()
```

### Now we can say that the restaurants which have book table facility their average rate is higher than the restaurants which doesnt't have this facility.
### so we should keep this fact in mind that if we have to open a restaurant then it should have table booking facility.

## 5.Now let's see the online order facility locationwise.

```python
df1 = food_df.groupby(['location','online_order'])['name'].count()
df1.to_csv('location_online.csv')
df1 = pd.read_csv('location_online.csv')
df1 = pd.pivot_table(df1, values=None, index=['location'], columns=['online_order'], fill_value=0, aggfunc=np.sum)
df1
```

```python
df1.plot(kind = 'bar', figsize = (15,8));
```

### From the above barplot we can easily understand that in areas like BTM, so many restaurants are already providing the online order facilty.
### so if we want to open a restaurant then we should open it in areas, where online order facility is not that much good and should provide online order facility so that our buisness will grow.

## 6. Now let's take a look in which areas restaurants are providing table booking facility.

```python
df2 = food_df.groupby(['location','book_table'])['name'].count()
df2.to_csv('location_booktable.csv')
df2 = pd.read_csv('location_booktable.csv')
df2 = pd.pivot_table(df2, values=None, index=['location'], columns=['book_table'], fill_value=0, aggfunc=np.sum)
df2
```

```python
df2.plot(kind = 'bar', figsize = (15,8));
```

### From the above plot we can say that most of the restaurants are not providing book table facility, but we should open a restaurant in the areas where the no. of restaurants are less and we can provide table booking facility in those areas so our competition will be less.

## 9. Now let's see the no. of votes according to the location

```python
df4 = food_df[['location', 'no_of_votes']]
df4.drop_duplicates()
df5 = df4.groupby(['location'])['no_of_votes'].sum()
df5 = df5.to_frame()
df5 = df5.sort_values('no_of_votes', ascending=False)
df5.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df5.index , df5['no_of_votes'])
plt.xticks(rotation = 90)
plt.show();
```

### From the above plot we can easily say that in which area no.votes are more,so that we should choose the areas like komanagala 5th block where we can get more customer feedback.

```python
food_df.head()
```

## 10. Now let's see which particular type of cuisine restaurant we should open?

```python
df6 = food_df[['cuisines', 'no_of_votes']]
df6.drop_duplicates()
df7 = df6.groupby(['cuisines'])['no_of_votes'].sum()
df7 = df7.to_frame()
df7 = df7.sort_values('no_of_votes', ascending=False)
df7.head()
```

### we can easily say that north indian cuisines restaurants are more popular.

```python
df7 = df7.iloc[1:, :]
df7.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df7.index , df7['no_of_votes'])
plt.xticks(rotation = 90)
plt.show();
```

### From the above plot we can say that north-indian,chinesse and south-indian cuisine are the top 3 cuisinies voted.
### so we should open a restaurant where we can provide North-indian,Chinesse and South-indian cuisines so it will be very easy for our restaurant to grow.
